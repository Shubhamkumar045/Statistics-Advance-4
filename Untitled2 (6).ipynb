```python
Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.
Answer--A t-test is used when the population standard deviation is unknown and must be estimated 
from the sample data.
It is appropriate for small sample sizes (typically less than 30) and when the population 
standard deviation is unknown.
There are two main types of t-tests: the one-sample t-test, used to compare a sample mean
to a known population mean, and the independent samples t-test, used to compare the means 
of two independent groups.
Example Scenario for T-Test:
Suppose a researcher wants to test whether there is a significant difference in the mean 
test scores of two groups of students who received different teaching methods. Since the 
population standard deviation is unknown and the sample sizes may not be large, the 
researcher would use an independent samples t-test to compare the means of the two groups.

Z-Test:

A z-test is used when the population standard deviation is known or when the sample 
size is large (typically greater than 30).
It is appropriate for situations where the population standard deviation is known 
and the sample size is sufficiently large to justify the use of the normal distribution.
The z-test is commonly used for hypothesis testing involving population proportions or means.
Example Scenario for Z-Test:
Consider a scenario where a quality control manager wants to determine if the mean
weight of boxes filled by a machine matches the specified mean weight of 100 grams.
The manager has data on weights from a large number of boxes and knows the population 
standard deviation. In this case, the manager can use a z-test to compare the sample
mean weight to the known population mean.

Q2: Differentiate between one-tailed and two-tailed tests.
Answer--One-tailed and two-tailed tests refer to the directionality of the hypothesis being tested and the 
corresponding critical region for rejection of the null hypothesis in hypothesis testing.

One-Tailed Test:

In a one-tailed test, the null hypothesis specifies the direction of the effect or difference being tested.
The alternative hypothesis asserts that the effect or difference exists in a specific direction.
The critical region for rejection of the null hypothesis is located entirely in one tail of the distribution.
One-tailed tests are used when there is a specific directional hypothesis or when the researcher is 
only interested in whether the effect is greater than or less than a certain value.
Example of One-Tailed Test:
Suppose a company claims that a new marketing strategy will increase sales by at least 10%. 
The null hypothesis (H0) would be that the new strategy has no effect or decreases sales, 
while the alternative hypothesis (H1) would be that the new strategy increases sales by at 
least 10%. In this case, the critical region for rejection of the null hypothesis would be
located in the upper tail of the distribution.

Two-Tailed Test:

In a two-tailed test, the null hypothesis does not specify the direction of the effect or difference being tested.
The alternative hypothesis asserts that the effect or difference exists, but it does not specify the direction.
The critical region for rejection of the null hypothesis is divided between both tails of the distribution.
Two-tailed tests are used when the researcher wants to determine whether there is a difference in either direction,
or when there is no specific directional hypothesis.
Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.
Answer--A Type I error occurs when the null hypothesis (H0) is incorrectly rejected when it is actually true.
It represents the situation where the researcher concludes that there is a significant effect or difference 
when, in reality, there is no effect or difference.
The probability of committing a Type I error is denoted by the significance level (α), which is the chosen
probability threshold for rejecting the null hypothesis.

Type II Error (False Negative):

A Type II error occurs when the null hypothesis (H0) is incorrectly retained (not rejected)
when the alternative hypothesis (H1) is actually true.
It represents the situation where the researcher fails to detect a significant effect or
difference that truly exists.
The probability of committing a Type II error is denoted by β.

Q4: Explain Bayes's theorem with an example.
Answer--Bayes' theorem is a fundamental concept in probability theory that describes how to update
the probability of a hypothesis based on new evidence. It allows us to calculate the probability
of a hypothesis given some observed data or evidence.

Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
Answer--A confidence interval is a range of values that is likely to contain the true value of a
population parameter, such as a population mean or proportion, with a certain level of confidence.
It provides a measure of the uncertainty or variability inherent in sample data.

Calculation of Confidence Interval:

The formula for calculating a confidence interval depends on the type of parameter being 
estimated (e.g., mean, proportion) and the distribution of the data. Here are the general
steps for calculating a confidence interval for the population mean (μ) using sample data:

Collect Sample Data: Obtain a sample from the population of interest. The sample should be
representative and collected using appropriate sampling techniques.

Calculate Sample Statistics: Calculate the sample mean (
�
ˉ
x
ˉ
 ) and sample standard deviation (s) from the sample data.

Choose Confidence Level: Select the desired confidence level (often 95%, 90%, or 99%). 
The confidence level represents the proportion of intervals, constructed using the same
method, that would contain the true population parameter if the process were repeated a
large number of times.

Determine Critical Value: Determine the critical value (z-score or t-score) corresponding to
the chosen confidence level and sample size. For large sample sizes (typically n > 30), the
z-score from the standard normal distribution is used. For small sample sizes (n < 30) or
when the population standard deviation is unknown, the t-score from the t-distribution is used.

Calculate Margin of Error: Calculate the margin of error, which represents the maximum 
likely difference between the sample estimate and the true population parameter.
The margin of error is determined by multiplying the critical value by the standard error of the sample mean.

Construct Confidence Interval: Use the sample mean and margin of error to construct
the confidence interval. The confidence interval is calculated as:
    
Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
event's probability and new evidence. Provide a sample problem and solution.
Answer--Suppose a certain medical test is known to correctly identify individuals with a certain disease 99% of the time (true positive rate) and correctly identify individuals without the disease 98% of the time (true negative rate).

Let's define the following events:

�
D: The event that a person has the disease.
+
+: The event that the person tests positive for the disease.
Given:

�
(
�
)
P(D), the prior probability of a person having the disease, is 0.01 (1% of the population has the disease).
�
(
+
∣
�
)
P(+∣D), the probability of testing positive given that the person has the disease, is 0.99 (true positive rate).
�
(
¬
�
)
P(¬D), the prior probability of a person not having the disease, is 0.99 (1 - 0.01).
�
(
+
∣
¬
�
)
P(+∣¬D), the probability of testing positive given that the person does not have the disease, is 0.02 (1 - true negative rate).
```
